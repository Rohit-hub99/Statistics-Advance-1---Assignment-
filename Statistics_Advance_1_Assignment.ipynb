# 1. What is a random variable in probability theory ?
  - A random variable in probability theory is a function that assigns a numerical value to each possible outcome of a random experiment.

  - Types of Random Variables:
    - Discrete Random Variable – Takes on a finite or countably infinite number of values.
      - Example: The number of heads when flipping three coins.
    - Continuous Random Variable – Takes on an uncountable number of values, usually within a range.
      - Example: The time a person waits at a bus stop.



# 2. What are the types of random variables ?
  - Random variables are classified into two main types:
    - Discrete Random Variables:
      - Take on countable values (finite or countably infinite).
      - Typically represent things like counts (e.g., number of heads in coin flips).
      - Described by Probability Mass Functions (PMF).
    - Examples:
      - Number of defective items in a batch.
      - Rolling a die (outcomes: 1, 2, 3, 4, 5, 6).

    - Continuous Random Variables:
      - Take on uncountable values (infinite, within an interval).
      - Typically represent measurements (e.g., time, temperature, height).
      - Described by Probability Density Functions (PDF).
    - Examples:
      - Time taken to finish a task.
      - Height of students in a class.

    - Special Types of Random Variables:
      - Bernoulli Variable – Takes only two values (0 or 1).
      - Binomial Variable – Sum of independent Bernoulli trials.
      - Poisson Variable – Counts events in a fixed interval (e.g., calls per minute).
      - Normal Variable – Follows a bell-shaped distribution (Gaussian).


# 3. What is the difference between discrete and continuous distributions ?
  - The difference between discrete and continuous distributions lies in the type of values their random variables take and how probabilities are assigned.

  - Discrete Distribution:
    - Random Variable Type - Countable values (0,1,2,3…)
    - Probability Function - PMF (Probability Mass Function)
    - Exact Value Probability	- Nonzero (P(X = x) can be positive).
    - Example Distributions	- Binomial, Poisson

  - Continuous Distribution:
    - Random Variable Type - Uncountable values (any real number in a range)
    - Probability Function - PDF (Probability Density Function)
    - Exact Value Probability	- Always Zero (P(X = x) = 0)
    - Example Distributions	- Normal, Exponential


# 4. What are probability distribution functions (PDF)?
  - A Probability Distribution Function (PDF) describes how probabilities are assigned to a random variable's possible values. There are two types based on the nature of the random variable:
  
  - Discrete Probability Distribution Function (PMF - Probability Mass Function)
    - Used for discrete random variables (countable values).
    - Gives the probability that a variable takes a specific value.

  - Continuous Probability Distribution Function (PDF - Probability Density Function)
    - Used for continuous random variables (uncountable values).
    - Describes the probability of values occurring within an interval.



# 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF) ?
  - A Cumulative Distribution Function (CDF) and a Probability Distribution Function (PDF) are both used to describe probabilities, but they have different purposes:

  - PDF (or PMF):
    - Definition:- Probability of a specific value (discrete) or density (continuous)
    - Representation:- PMF for discrete, PDF for continuous
    - Probability at a Single Point:-	Nonzero for discrete, Zero for continuous
    - Graph:-	Bell-shaped (normal), spike-like (discrete)

  - CDF:
    - Definition:- Probability of X is less than equal to x
    - Representation:- Summation (discrete) or Integral (continuous)
    - Probability at a Single Point:- Always increases as x increases
    - Graph:- Always non-decreasing, S-shaped for normal

# 11. What are the characteristics of a normal distribution ?
  - The Normal Distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric around its mean. It is one of the most important distributions in statistics, used to model natural phenomena like human heights, exam scores, and stock returns.

  - Key Characteristics of a Normal Distribution:
    - Symmetric about the mean (μ).
    - Mean, median, and mode are equal.
    - Bell-shaped curve.
    - Defined by μ (location) and σ power 2 (spread).
    - Total probability = 1.
    - Empirical Rule (68-95-99.7 Rule):
      - 68% of values fall within 1 standard deviation of the mean (μ ± σ).
      - 95% of values fall within 2 standard deviations (μ ± 2σ).
      - 99.7% of values fall within 3 standard deviations (μ ± 3σ).



# 12. What is the standard normal distribution, and why is it important ?
  - The Standard Normal Distribution is a special case of the Normal Distribution where:
    - The mean μ = 0
    - The standard deviation σ = 1
    - It is denoted as:
      - Z∼N(0,1)
    - It is also called the Z-distribution because it is used for Z-scores and standardization.

  - Importance of Standard Normal Distribution:
    - Makes Any Normal Distribution Comparable:
      - Many real-world data follow a normal distribution, but they have different means (μ) and standard deviations (σ)
      - By converting values into Z-scores, we can compare them across different normal distributions.
    
    - Simplifies Probability Calculations:
      - The standard normal table (Z-table) provides probabilities for any normal distribution without complex integration.
      - Instead of computing probabilities for different means and standard deviations, we standardize to N(0,1).
    
    - Used in Hypothesis Testing & Confidence Intervals:
      - The standard normal distribution is the foundation for Z-tests, which are used in statistical inference.
      - It helps construct confidence intervals for means and proportions.
    
    - Empirical Rule (68-95-99.7 Rule):
      - Since N(0,1) N(0,1) follows the normal curve, we can use this rule for quick probability approximations:
        - 68% of values are within 1 standard deviation of the mean.
        - 95% are within 2 standard deviations.
        - 99.7% are within 3 standard deviations.


# 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics ?
  - The Central Limit Theorem (CLT) is one of the most fundamental concepts in statistics. It states that:
  - Regardless of the original distribution of a population, the distribution of the sample mean will approximate a normal distribution as the sample size increases.
  - In simpler terms:
    - If you take many random samples from any population (no matter the shape of its distribution),
    - And calculate their means,\
    - Then the distribution of those means will become approximately normal, even if the population itself is not normal,
    - This approximation improves as the sample size n increases (typically n ≥ 30)

  - Why is the CLT Important:
    - Enables Normal Approximation
      - Even if a population is skewed or non-normal, we can assume the sample means are normally distributed for large n.
      - This allows us to apply normal distribution techniques (e.g., confidence intervals, hypothesis tests) to non-normal data.

    - Foundation for Hypothesis Testing
      - Many statistical tests (e.g., t-tests, ANOVA, regression) rely on the assumption that sample means are normally distributed.
      - CLT justifies their use even when the original population is not normal.
    
    - Allows Sampling-Based Estimation
      - Since collecting entire population data is often impractical, CLT lets us make inferences about a population using a small sample.



# 14. How does the Central Limit Theorem relate to the normal distribution?
  - The Central Limit Theorem (CLT) explains why normal distribution is so commonly observed in real-world data and why it's the foundation of statistical inference.

  - CLT Leads to Normality of Sample Means
    - If we take random samples from any population (whether normal, skewed, or uniform),
    - And compute their sample means,
    - The distribution of these sample means will approach a normal distribution, regardless of the population's shape.
    - This happens as the sample size (n) increases.

  - The Normal Distribution is a Special Case of CLT
    - If the original population is already normal, then the sample means will also be normally distributed for any sample size.
    - But even if the original population isn't normal, the sample means will still become normal if n is large enough.

  - CLT Justifies Using the Normal Distribution in Statistics
    - We can estimate population parameters (like the mean) with a normal-based margin of error.
    - Since sample means are normally distributed, we can calculate probabilities and make inferences.
    - Many algorithms assume normality in their data (e.g., linear regression residuals).

# 16. How do you calculate a Z-score, and what does it represent ?
  - A Z-score (also called a standard score) measures how many standard deviations a data point is from the mean of a distribution.

  - Formula for Z-Score:
    - For a single value X:
        Z = X - μ / σ
    
    - For a sample mean X bar:
       Z = sample mean - population mean / population standard deviation / Square root of sample size

   - What Does a Z-Score Represent?
     - Indicates How Unusual a Value Is:
       - Z=0 → The value is exactly at the mean.
       - Z=1 → The value is 1 standard deviation above the mean.
       - Z=−1 → The value is 1 standard deviation below the mean.
       - ∣Z∣>2 → The value is unusual.
       - ∣Z∣>3 → The value is very rare (in a normal distribution).
     - Used for Probability Calculations:
       - The Z-table provides the probability of a value occurring in a normal distribution.
       - Example: If Z=1.96, about 97.5% of values are below it.
     - Standardization for Comparisons:
       - Converts different distributions into a common scale (mean = 0, standard deviation = 1).
       - Example: Comparing SAT and GRE scores.



# 17.  What are point estimates and interval estimates in statistics ?
  - Point Estimates:
     - A point estimate is a single value that serves as an approximation of a population parameter.
       - Examples of Point Estimates:
         - Sample mean (X bar) estimates the population mean (μ)
         - Sample proportion  (p^) estimates the population proportion (p)
         - Sample standard deviation (s) estimates the population standard deviation (σ)

      - Interval Estimates (Confidence Intervals)
        - An interval estimate provides a range of values within which the population parameter is likely to fall.



# 18. What is the significance of confidence intervals in statistical analysis ?
  - A Confidence Interval (CI) is a range of values used to estimate an unknown population parameter (like mean or proportion) with a specified level of confidence.

  - Why Are Confidence Intervals Important:
     - Measure of Uncertainty
       - A CI quantifies the uncertainty in estimating a population parameter.
       - Instead of relying on a single value (point estimate), it provides a range that accounts for sample variability.
     - More Informative than Point Estimates
       - A sample mean alone is not very useful without knowing how much it might vary
       - A CI gives a range, indicating the precision of the estimate.
       - Example: Saying "The average height is 165 cm" is less informative than "The average height is 165 cm ± 3 cm (95% CI: 162 to 168 cm)".
     - Used in Hypothesis Testing
       - If a CI excludes a hypothesized value (e.g.μ0 =170), we reject the null hypothesis.
       - If a CI includes the hypothesized value, we fail to reject the null hypothesis.
     - Helps in Decision Making
       - In medical research, CI helps assess drug effectiveness
       - In business, CI helps in market research predictions.
       - In engineering, CI helps evaluate process stability.



# 19. What is the relationship between a Z-score and a confidence interval ?
  - A Z-score and a confidence interval (CI) are closely related because the Z-score determines how wide the confidence interval will be.

  - How Are Z-Scores and Confidence Intervals Related:
    - The Z-score represents the number of standard deviations a value is from the mean.
    - In confidence intervals, the Z-score determines the margin of error (MOE), which affects the range of the interval.
    - Confidence intervals estimate population parameters with a specified level of certainty.
    - A wider confidence interval means greater uncertainty, but more confidence in capturing the true parameter.

# 21. What are the assumptions for applying the Central Limit Theorem ?
  - The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the shape of the original population distribution.

  - Assumptions for the CLT to Hold:
    - Random Sampling
      - The sample must be randomly selected from the population.
      - If the sampling process is biased, the CLT may not apply.

    - Sample Size Should Be Large Enough
      - Rule of Thumb: n≥30 is generally sufficient.
      - If the population is normal, even a small sample size (e.g., n=5) is enough.
      - If the population is skewed or non-normal, larger sample sizes(n>30 or more) are required.

    - Independence of Observations
      - Observations must be independent (one data point should not influence another).
      - If data points are dependent (e.g., time series, clustered data), CLT may not work properly.

    -  Finite Variance (σ power 2<∞)
      - The population should have a finite standard deviation.
      - If the data has extremely high variance or infinite variance, CLT does not hold (e.g., Cauchy distribution).



# 22. What is the concept of expected value in a probability distribution ?
  - The expected value (EV) of a random variable is the long-term average of outcomes in a probability distribution. It represents the theoretical mean of the distribution, assuming the experiment is repeated infinitely.

  - Definition:
    - For a discrete or continuous probability distribution, the expected value is a weighted average of all possible values, where the weights are the probabilities of each value occurring.


# 23. How does a probability distribution relate to the expected outcome of a random variable ?
  - A probability distribution describes how the values of a random variable are distributed, specifying the probabilities of different outcomes. The expected outcome (expected value, E(X) is the long-run average value of the random variable when repeated many times.

  - Key Relationship:
    - The expected value is computed by weighting each possible outcome by its probability and summing (discrete case) or integrating (continuous case).
  
  - Symmetric Distributions (e.g., Normal, Uniform):
    - The expected value is at the center (mean = median = mode).

  - Skewed Distributions (e.g., Exponential, Log-normal):
    - The expected value shifts toward the longer tail.
    - Example: In a right-skewed distribution, the expected value is greater than the median.
  
  - Multimodal Distributions:
    - Expected value may be between peaks but may not represent a likely value.

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import random
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from collections import Counter

def generate_discrete_uniform(n, a, b):
    """Generate n samples from a discrete uniform distribution U(a, b)"""
    return [random.randint(a, b) for _ in range(n)]

# Parameters
a, b = 1, 6  # Simulating a fair 6-sided die
n_samples = 10000  # Number of samples

# Generate samples
samples = generate_discrete_uniform(n_samples, a, b)

# Compute PMF
counts = Counter(samples)
k_values = sorted(counts.keys())
p_values = [counts[k] / n_samples for k in k_values]

# Plot PMF
plt.figure(figsize=(8, 5))
sns.barplot(x=k_values, y=p_values, color='blue', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF) of Discrete Uniform Distribution")
plt.ylim(0, max(p_values) + 0.01)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution

import random
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from collections import Counter
from scipy.stats import bernoulli, binom, poisson

def generate_discrete_uniform(n, a, b):
    """Generate n samples from a discrete uniform distribution U(a, b)"""
    return [random.randint(a, b) for _ in range(n)]

def bernoulli_pdf(p, x):
    """Calculate the probability mass function (PMF) of a Bernoulli distribution"""
    if x not in [0, 1]:
        return 0  # Bernoulli variable can only be 0 or 1
    return p if x == 1 else (1 - p)

def simulate_binomial(n, p, size):
    """Simulate a binomial distribution with parameters n and p"""
    return np.random.binomial(n, p, size)

def simulate_poisson(lam, size):
    """Simulate a Poisson distribution with parameter lambda"""
    return np.random.poisson(lam, size)

# Parameters for binomial distribution
n, p, size = 10, 0.5, 10000

# Generate binomial samples
binomial_samples = simulate_binomial(n, p, size)

# Plot histogram for binomial distribution
plt.figure(figsize=(8, 5))
sns.histplot(binomial_samples, bins=n+1, kde=False, color='blue', edgecolor='black', stat='probability')
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Histogram of Binomial Distribution (n=10, p=0.5)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Parameters for Poisson distribution
lam = 4  # Average rate of occurrences

# Generate Poisson samples
poisson_samples = simulate_poisson(lam, size)

# Plot histogram for Poisson distribution
plt.figure(figsize=(8, 5))
sns.histplot(poisson_samples, bins=range(0, max(poisson_samples)+1), kde=False, color='red', edgecolor='black', stat='probability')
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.title("Histogram of Poisson Distribution (λ=4)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Parameters for discrete uniform distribution
a, b = 1, 6  # Simulating a fair 6-sided die
n_samples = 10000  # Number of samples

# Generate samples
samples = generate_discrete_uniform(n_samples, a, b)

# Compute PMF
counts = Counter(samples)
k_values = sorted(counts.keys())
p_values = [counts[k] / n_samples for k in k_values]

# Compute CDF
cdf_values = np.cumsum(p_values)

# Plot PMF for discrete uniform distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=k_values, y=p_values, color='blue', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF) of Discrete Uniform Distribution")
plt.ylim(0, max(p_values) + 0.01)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot CDF for discrete uniform distribution
plt.figure(figsize=(8, 5))
plt.step(k_values, cdf_values, where='mid', color='green', linewidth=2)
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

import numpy as np

# Step 1: Generate a normally distributed dataset
np.random.seed(42)  # For reproducibility
mu, sigma = 50, 10  # Mean and standard deviation
n = 1000  # Sample size
data = np.random.normal(mu, sigma, n)  # Generate data

# Step 2: Compute Range
data_range = np.max(data) - np.min(data)

# Step 3: Compute Interquartile Range (IQR)
Q1 = np.percentile(data, 25)  # 25th percentile
Q3 = np.percentile(data, 75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile Range

# Print results
print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {IQR:.2f}")
```

```python
# 27.  Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Step 1: Generate a normally distributed dataset
np.random.seed(42)  # For reproducibility
mu, sigma = 50, 15  # Mean and standard deviation
n = 1000  # Sample size
data = np.random.normal(mu, sigma, n)  # Generate data

# Step 2: Apply Z-score normalization
normalized_data = zscore(data)  # Z-score transformation

# Step 3: Plot original and normalized distributions
plt.figure(figsize=(10, 5))

# Original Data Histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label="Mean")
plt.xlabel('Original Values')
plt.ylabel('Frequency')
plt.title('Original Data Distribution')
plt.legend()
plt.grid(True)

# Normalized Data Histogram
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, alpha=0.7, color='green', edgecolor='black')
plt.axvline(np.mean(normalized_data), color='red', linestyle='dashed', linewidth=2, label="Mean (0)")
plt.xlabel('Z-Score Values')
plt.title('Z-Score Normalized Distribution')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Print summary statistics
print(f"Original Data: Mean = {np.mean(data):.2f}, Std Dev = {np.std(data):.2f}")
print(f"Normalized Data: Mean = {np.mean(normalized_data):.2f}, Std Dev = {np.std(normalized_data):.2f}")
```

```python
# 28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(data):
    """
    Computes the skewness and kurtosis of a dataset.

    Parameters:
    data (array-like): Input dataset.

    Returns:
    tuple: (skewness, kurtosis)
    """
    skewness = skew(data)
    excess_kurtosis = kurtosis(data)  # Excess kurtosis (normal = 0)
    return skewness, excess_kurtosis

# Generate normally distributed data
np.random.seed(42)
mu, sigma = 50, 10  # Mean and standard deviation
n = 1000  # Sample size
data = np.random.normal(mu, sigma, n)

# Compute skewness and kurtosis
skewness, excess_kurtosis = compute_skewness_kurtosis(data)

# Print results
print(f"Skewness: {skewness:.4f}")
print(f"Excess Kurtosis: {excess_kurtosis:.4f} (Total Kurtosis = {excess_kurtosis + 3:.4f})")
```
